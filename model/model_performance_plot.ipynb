```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
plt.style.use("ggplot")
%matplotlib inline
```

```python
data = {
    'nn - 2 body - same mass': [0.6736981971991173, 0.7087378867937281, 1.2318787586011237],
    'gb - 2 body - same mass': [0.21754216531914175, 0.4091465745889179, 0.6803796973228321],
    'gb - 3 body - same mass': [0.44162589578527767, 0.6883552137643727, 1.23045057770961],
    'gb - 3 body - diff mass': [0.926865444562274, 1.3189722859293307, 2.342292422475718],
    'gb - 2 body - same mass - optimized': [0, 0, 0],
    'gb - 3 body - same mass - optimized': [0, 0, 0],
}

df = pd.DataFrame(data, index=['10','100','1000'])
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12,8))

for col in df.columns:
    ax.plot(df[col].index, df[col].values, lw=3, marker='o', ms=10, label=col)

ax.legend(fontsize=13)
ax.set_title('Model Performance for Various Time Horizons', fontsize=20)
ax.set_xlabel('Time Steps of Prediction After Initial Position', fontsize=15)
ax.set_ylabel('Root Mean Squared Error', fontsize=15)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(15)

fig.savefig('../images/model_perf.png', bbox_inches='tight')
```
